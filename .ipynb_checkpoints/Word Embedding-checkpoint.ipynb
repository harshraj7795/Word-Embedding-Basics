```python
#importing libraries for one hot encoding
from tensorflow.keras.preprocessing.text import one_hot
```

```python
txt=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
#size of vocabulary
vocab=10000
```

```python
#representing the sentences as one hot encoding
onehot_rep=[one_hot(words,vocab) for words in txt]
onehot_rep
```

```python
#importing libraries for embedding words
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
#padding the onehot rep for obtaining consistent dimensions in matrix while embedding
sent_length=8
padded_mat=pad_sequences(onehot_rep,padding='pre',maxlen=sent_length)
print(padded_mat)
```

```python
#preparing the model by applying embedding
dim=10   #dimension for embedding
model = Sequential()
model.add(Embedding(vocab,dim,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(padded_mat)
```
